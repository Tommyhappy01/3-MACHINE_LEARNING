## Evaluation on the Test Set

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = model_poly.predict(X_test)
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```

```python
y_train_pred = model_poly.predict(X_train)
```

```python
eval_metric(y_train, y_train_pred)
```

## Let's find optimal degree of poly

```python
def poly(d):
    
    train_rmse_errors = []
    test_rmse_errors = []
    
    for i in range(1, d):
        polynomial_converter = PolynomialFeatures(degree = i, include_bias =False)
        poly_features = polynomial_converter.fit_transform(X)
        
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
        
        model = LinearRegression(fit_intercept=True)
        model.fit(X_train, y_train)
        
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
        test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
        
        train_rmse_errors.append(train_RMSE)
        test_rmse_errors.append(test_RMSE)
        
    return pd.DataFrame({"train_rmse_errors": train_rmse_errors, "test_rmse_errors":test_rmse_errors}, index=[i for i in range(1,d)])
        
```

Polinom linear reg. ite tamamen skorlarmz iyiletirir. Ama feature saymz artacak, hangi birini mteriye anlatacaz ki. 1000 tane feature mz olacak. Yani bunu insight etmek, mteriye aklamak anlamsz. Sadece predictimizin baarsn sentetik veriler ile arttrmaya yaryor.
Train ve test setimizin sonular arasndaki uurum overfitting olduunu gsterir. (edited) 

```python
poly(10)
```

```python
plt.plot(range(1,10), poly(10)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,10), poly(10)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
```

```python
plt.plot(range(1,6), poly(6)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,6), poly(6)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
```

2'yi ve 3' semede sknt yoktur. Ama patlama 4 te balad iin 4 noktas seilemez. nk 4 noktas en salam en kuvvetli en nemli feature olsa bile bu nokta overfitting in balad noktadr. (edited)
